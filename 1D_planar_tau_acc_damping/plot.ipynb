```python
import matplotlib.pyplot as plt
import numpy as np
import csv

p = np.loadtxt("p.csv", dtype=float)
f0_1 = np.loadtxt("f0_1.csv", dtype=float)
f0_2 = np.loadtxt("f0_2.csv", dtype=float)

# plot

fig, ax = plt.subplots()
fig.patch.set_facecolor('white')
ax.set_facecolor('white')

ax.plot(p*3e10/1.6e-3, np.power(p,4)*f0_1, ls='-')
ax.plot(p*3e10/1.6e-3, np.power(p,4)*f0_2, ls='-')

ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlim(1e0,1e7)
# ax.set_ylim(1e-1,2e0)
ax.set_xlabel("p (GeV/c)")
ax.set_ylabel(r"$\rm p^{4} \, f(p)$")
plt.tight_layout()
# plt.savefig("n0.png", dpi=500)
plt.show()
```
