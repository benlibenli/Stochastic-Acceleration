```python
import matplotlib.pyplot as plt
import numpy as np
import csv

p = np.loadtxt("p.csv", dtype=float)
Dz_i = np.loadtxt("Dz_i.csv", dtype=float)
Dz_f = np.loadtxt("Dz_f.csv", dtype=float)
f0_1 = np.loadtxt("f0_1.csv", dtype=float)
f0_2 = np.loadtxt("f0_2.csv", dtype=float)
f_1 = np.loadtxt("f_1.csv", dtype=float)
f_2 = np.loadtxt("f_2.csv", dtype=float)

# plot 1

fig, ax = plt.subplots()

ax.plot(p*3e10/1.6e-3, np.power(p,4)*f0_1, ls='-', c="black", label="1st order (TS)")
ax.plot(p*3e10/1.6e-3, np.power(p,4)*f0_2, ls='-', c="red", label="1st order & 2nd order (TS)")

ax.plot(p*3e10/1.6e-3, np.power(p,4)*f_1, ls='--', c="black", label="1st order (0.5*z0)")
ax.plot(p*3e10/1.6e-3, np.power(p,4)*f_2, ls='--', c="red", label="1st order & 2nd order (0.5*z0)")

ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlim(1e0,1e7)
ax.set_ylim(2e-24,1e-23)
ax.set_xlabel("p (GeV/c)")
ax.set_ylabel(r"$\rm p^{4} \, f(p)$")
ax.set_title("No damping")
ax.legend(loc="lower left")
plt.tight_layout()
plt.savefig("f.png", dpi=500)
plt.show()

# plot 2

fig, ax = plt.subplots()

ax.plot(p*3e10/1.6e-3, Dz_i, ls="-", c="black", label=r"$D(p)\propto p^{1/3}$")
ax.plot(p*3e10/1.6e-3, Dz_f, ls="--", c="red", label="after 1st update")

ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlim(1e0,1e7)
ax.set_ylim(1e25,1e35)
ax.set_xlabel("p (GeV/c)")
ax.set_ylabel(r"$\rm D(p)$")
ax.legend()
plt.tight_layout()
plt.savefig("D.png", dpi=500)
plt.show()
```
